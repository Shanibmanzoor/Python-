```python
import random
import time
```

```python
rock = 1
paper = 2
scissors = 3
```

```python
name = {
    rock: "Rock",
    paper: "Paper",
    scissors: "Scissors"
}
```

```python
rules = {
    rock: scissors,
    paper: rock,
    scissors: paper
}
```

```python
player_score = 0
computer_score = 0
```

```python
def start():
  print("Let's play Rock Paper Scissors")
  print("------------------------------")
  while game():
   pass
  scores()
```

```python
def game():
  player = move()
  computer = random.randint(1,3)
  print(f"you chose {name[player]}")
  print(f"computer chose {name[computer]}")
  result(player,computer)
  return play_again()
```

```python
def move():
  while True:
    player = input("Rock = 1\nPaper = 2\nScissors = 3\n")
    try:
      player = int(player)
      if player in (1,2,3):
        return player
    except ValueError:
      pass
      print("Oops! I didn't understand that. Try again.")
```

```python
def result(player,computer):
 print("1---------")
 time.sleep(0.5)
 print("2---------")
 time.sleep(0.5)
 print("3---------")
 time.sleep(0.5)

 print(f"computer threw:- {name[computer]}")
 global player_score
 global computer_score

 if player == computer:
    print("tie game")
 elif rules[player] == computer:
    print("you win")
    player_score += 1
 else:
    print("you lose")
    computer_score += 1
```

```python
def play_again():
  answer = input("Would you like to play again? y/n: ")
  if answer in ("y","Y","yes","Yes"):
    return answer
  else:
    print("Thank you very much for playing our game.")
```

```python
def scores():
  global player_score
  global computer_score
  print(f"your score:- {player_score}")
  print(f"computer score:- {computer_score}")
```
